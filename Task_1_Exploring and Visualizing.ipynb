```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
print("Libraries imported successfully!")
print(f"pandas  version: {pd.__version__}")
print(f"seaborn version: {sns.__version__}")
```

```python
# Load Iris dataset using seaborn (no download needed)
df = sns.load_dataset("iris")

print("Dataset loaded successfully!")
print(f"Shape         : {df.shape}  (rows x columns)")
print(f"Column names  : {list(df.columns)}")
print()
print("=== First 10 Rows ===")
df.head(10)
```

```python
print("=== .info() — Data Types & Memory ===")
df.info()
print()
print("=== Missing Values per Column ===")
print(df.isnull().sum())
print()
print("=== Species Class Distribution ===")
print(df["species"].value_counts())
```

```python
print("=== .describe() — Descriptive Statistics ===")
df.describe()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Sepal dimensions
sns.scatterplot(ax=axes[0], data=df, x="sepal_length", y="sepal_width",
                hue="species", palette="Set1", s=80, edgecolor="white")
axes[0].set_title("Sepal Length vs Sepal Width", fontsize=13, fontweight="bold")
axes[0].set_xlabel("Sepal Length (cm)")
axes[0].set_ylabel("Sepal Width (cm)")

# Plot 2: Petal dimensions
sns.scatterplot(ax=axes[1], data=df, x="petal_length", y="petal_width",
                hue="species", palette="Set1", s=80, edgecolor="white")
axes[1].set_title("Petal Length vs Petal Width", fontsize=13, fontweight="bold")
axes[1].set_xlabel("Petal Length (cm)")
axes[1].set_ylabel("Petal Width (cm)")

plt.suptitle("Iris Dataset — Scatter Plots by Species", fontsize=15, fontweight="bold")
plt.tight_layout()
plt.savefig("scatter_plots.png", dpi=100, bbox_inches="tight")
plt.show()
print("Observation: Petal features clearly separate species better than sepal features!")
```

```python
features = ["sepal_length", "sepal_width", "petal_length", "petal_width"]
species_colors = {"setosa": "#e74c3c", "versicolor": "#3498db", "virginica": "#2ecc71"}

fig, axes = plt.subplots(2, 2, figsize=(13, 9))
axes = axes.flatten()

for i, feat in enumerate(features):
    for species, color in species_colors.items():
        subset = df[df["species"] == species][feat]
        axes[i].hist(subset, bins=15, alpha=0.65, color=color,
                     label=species, edgecolor="white")
    axes[i].set_title(f"Distribution of {feat.replace('_', ' ').title()}",
                      fontsize=12, fontweight="bold")
    axes[i].set_xlabel(f"{feat} (cm)")
    axes[i].set_ylabel("Frequency")
    axes[i].legend()

plt.suptitle("Histograms — Iris Feature Distributions by Species",
             fontsize=14, fontweight="bold")
plt.tight_layout()
plt.savefig("histograms.png", dpi=100, bbox_inches="tight")
plt.show()
print("Observation: Petal length/width shows the clearest species separation.")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(13, 9))
axes = axes.flatten()

for i, feat in enumerate(features):
    sns.boxplot(ax=axes[i], data=df, x="species", y=feat,
                palette="Set2", width=0.5,
                flierprops=dict(marker="o", color="red", markersize=6, alpha=0.6))
    axes[i].set_title(f"Box Plot: {feat.replace('_', ' ').title()}",
                      fontsize=12, fontweight="bold")
    axes[i].set_xlabel("Species")
    axes[i].set_ylabel(f"{feat} (cm)")

plt.suptitle("Box Plots — Outlier Detection by Feature & Species",
             fontsize=14, fontweight="bold")
plt.tight_layout()
plt.savefig("boxplots.png", dpi=100, bbox_inches="tight")
plt.show()
print("Observation: Setosa has notably smaller petal dimensions. Some sepal_width outliers visible.")
```

```python
pair = sns.pairplot(df, hue="species", palette="Set1",
                    diag_kind="kde", height=2.4,
                    plot_kws={"alpha": 0.6, "edgecolor": "none"})
pair.fig.suptitle("Pair Plot — All Iris Features", y=1.02,
                   fontsize=14, fontweight="bold")
plt.savefig("pairplot.png", dpi=100, bbox_inches="tight")
plt.show()
print("Pair plot saved — shows all pairwise relationships!")
```

```python
plt.figure(figsize=(8, 6))
corr = df[features].corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm",
            linewidths=0.5, square=True,
            cbar_kws={"shrink": 0.75},
            annot_kws={"size": 12})
plt.title("Feature Correlation Heatmap", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.savefig("heatmap.png", dpi=100, bbox_inches="tight")
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(13, 9))
axes = axes.flatten()

for i, feat in enumerate(features):
    sns.violinplot(ax=axes[i], data=df, x="species", y=feat,
                   palette="muted", inner="quartile")
    axes[i].set_title(f"Violin Plot: {feat.replace('_', ' ').title()}",
                      fontsize=12, fontweight="bold")
    axes[i].set_xlabel("Species")

plt.suptitle("Violin Plots — Feature Distribution Shape by Species",
             fontsize=14, fontweight="bold")
plt.tight_layout()
plt.savefig("violin_plots.png", dpi=100, bbox_inches="tight")
plt.show()
```

```python
print("=== Summary Statistics by Species ===")
print()
summary = df.groupby("species")[features].agg(["mean", "std", "min", "max"])
print(summary.round(2))
```
